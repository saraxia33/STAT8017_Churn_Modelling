```python
from sklearn.datasets import load_iris
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
```

```python
df = pd.read_csv("Churn_Modelling.csv", index_col=[1])
del df['RowNumber']
del df['Surname']

df.Gender[df.Gender == 'Male'] = 1
df.Gender[df.Gender == 'Female'] = 0

expl = ['CreditScore', 'Gender', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'EstimatedSalary']
x = df[expl]
y = df[['Exited']]


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

Testing accuracy at the bottom of the notebook.

```python
activation = ['identity', 'relu', 'tanh', 'logistic']
acc = []
for act in activation:
    clf = MLPClassifier(hidden_layer_sizes = [5], activation=act, solver='adam', learning_rate_init = 0.01, max_iter = 1000, verbose = True)
    clf.fit(x_train, y_train)
    acc.append(accuracy_score(y_test, clf.predict(x_test)))
    print("Testing Accuracy:", acc[-1])
    
```

```python
for i in range(4):
    print('Testing Accuracy with ', activation[i], ': ', acc[i])
```
